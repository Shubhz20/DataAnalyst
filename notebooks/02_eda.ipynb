## Exploratory Data Analysis (EDA)

### 1. Descriptive Statistics

Let's start by looking at the descriptive statistics for both numerical and categorical columns to get an overview of the data.

```python
print("Descriptive Statistics for Numerical Columns:")
display(df_cleaned.describe())

print("\nDescriptive Statistics for Categorical Columns:")
display(df_cleaned.describe(include='object'))
```

### 2. Univariate Analysis

We will visualize the distribution of key numerical and categorical variables.

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

# Histograms for numerical columns
fig, axes = plt.subplots(3, 2, figsize=(16, 18))
fig.suptitle('Distribution of Key Numerical Variables', fontsize=16)

sns.histplot(df_cleaned['Cost Price'], kde=True, ax=axes[0, 0], bins=20)
axes[0, 0].set_title('Distribution of Cost Price')

sns.histplot(df_cleaned['Retail Price'], kde=True, ax=axes[0, 1], bins=20)
axes[0, 1].set_title('Distribution of Retail Price')

sns.histplot(df_cleaned['Profit Margin'], kde=True, ax=axes[1, 0], bins=20)
axes[1, 0].set_title('Distribution of Profit Margin')

sns.histplot(df_cleaned['Order Quantity'], kde=True, ax=axes[1, 1], bins=15)
axes[1, 1].set_title('Distribution of Order Quantity')

sns.histplot(df_cleaned['Total'], kde=True, ax=axes[2, 0], bins=25)
axes[2, 0].set_title('Distribution of Total Amount')

# Hide unused subplot
axes[2, 1].axis('off')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
# Count plots for categorical columns
fig, axes = plt.subplots(2, 2, figsize=(18, 14))
fig.suptitle('Distribution of Key Categorical Variables', fontsize=16)

sns.countplot(x='Customer Type', data=df_cleaned, ax=axes[0, 0], palette='viridis', order=df_cleaned['Customer Type'].value_counts().index)
axes[0, 0].set_title('Distribution of Customer Type')
axes[0, 0].set_xlabel('Customer Type')
axes[0, 0].set_ylabel('Count')
axes[0, 0].tick_params(axis='x', rotation=45)

sns.countplot(x='Order Priority', data=df_cleaned, ax=axes[0, 1], palette='magma', order=df_cleaned['Order Priority'].value_counts().index)
axes[0, 1].set_title('Distribution of Order Priority')
axes[0, 1].set_xlabel('Order Priority')
axes[0, 1].set_ylabel('Count')
axes[0, 1].tick_params(axis='x', rotation=45)

sns.countplot(x='Product Category', data=df_cleaned, ax=axes[1, 0], palette='crest', order=df_cleaned['Product Category'].value_counts().index)
axes[1, 0].set_title('Distribution of Product Category')
axes[1, 0].set_xlabel('Product Category')
axes[1, 0].set_ylabel('Count')
axes[1, 0].tick_params(axis='x', rotation=45)

sns.countplot(x='Ship Mode', data=df_cleaned, ax=axes[1, 1], palette='flare', order=df_cleaned['Ship Mode'].value_counts().index)
axes[1, 1].set_title('Distribution of Ship Mode')
axes[1, 1].set_xlabel('Ship Mode')
axes[1, 1].set_ylabel('Count')
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

### 3. Bivariate Analysis

Now, let's explore relationships between different variables. We'll examine 'Total' against 'Product Category', 'Customer Type', and 'Order Quantity'.

```python
fig, axes = plt.subplots(1, 3, figsize=(22, 7))
fig.suptitle('Bivariate Analysis of Sales', fontsize=16)

# Total by Product Category
sns.boxplot(x='Product Category', y='Total', data=df_cleaned, ax=axes[0], palette='viridis')
axes[0].set_title('Total by Product Category')
axes[0].set_xlabel('Product Category')
axes[0].set_ylabel('Total Amount')
axes[0].tick_params(axis='x', rotation=45)

# Total by Customer Type
sns.boxplot(x='Customer Type', y='Total', data=df_cleaned, ax=axes[1], palette='magma')
axes[1].set_title('Total by Customer Type')
axes[1].set_xlabel('Customer Type')
axes[1].set_ylabel('Total Amount')
axes[1].tick_params(axis='x', rotation=45)

# Total vs. Order Quantity (using a scatter plot or regression plot)
sns.regplot(x='Order Quantity', y='Total', data=df_cleaned, ax=axes[2], scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
axes[2].set_title('Total vs. Order Quantity')
axes[2].set_xlabel('Order Quantity')
axes[2].set_ylabel('Total Amount')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
daily_sales = df_cleaned.groupby('Order Date')['Total'].sum().reset_index()

plt.figure(figsize=(15, 7))
sns.lineplot(x='Order Date', y='Total', data=daily_sales)
plt.title('Daily Total Sales Trend Over Time', fontsize=16)
plt.xlabel('Order Date')
plt.ylabel('Total Sales Amount')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
numerical_cols = df_cleaned.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df_cleaned[numerical_cols].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Variables', fontsize=16)
plt.tight_layout()
plt.show()
```
