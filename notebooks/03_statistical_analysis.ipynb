## Exploratory Data Analysis (EDA)

### 1. Descriptive Statistics

Let's start by looking at the descriptive statistics for both numerical and categorical columns to get an overview of the data.

```python
print("Descriptive Statistics for Numerical Columns:")
display(df_cleaned.describe())

print("\nDescriptive Statistics for Categorical Columns:")
display(df_cleaned.describe(include='object'))
```

### 1. ANOVA: Comparing 'Total' Sales Across 'Product Category'

We will use ANOVA to test if there are statistically significant differences in the mean 'Total' sales among different 'Product Categories'.

```python
from scipy.stats import f_oneway

# Get unique product categories
product_categories = df_cleaned['Product Category'].unique()

# Create a list of 'Total' sales for each category
sales_by_category = [df_cleaned[df_cleaned['Product Category'] == category]['Total'] for category in product_categories]

# Perform ANOVA
f_statistic, p_value = f_oneway(*sales_by_category)

print(f"ANOVA F-statistic for Product Category: {f_statistic:.2f}")
print(f"ANOVA p-value for Product Category: {p_value:.3f}")

if p_value < 0.05:
    print("\nConclusion: There are statistically significant differences in mean total sales across different product categories.")
else:
    print("\nConclusion: There are no statistically significant differences in mean total sales across different product categories.")
```

### 2. ANOVA: Comparing 'Total' Sales Across 'Customer Type'

Similarly, let's use ANOVA to test for significant differences in mean 'Total' sales among different 'Customer Types'.

```python
from scipy.stats import f_oneway

# Get unique customer types
customer_types = df_cleaned['Customer Type'].unique()

# Create a list of 'Total' sales for each customer type
sales_by_customer_type = [df_cleaned[df_cleaned['Customer Type'] == cust_type]['Total'] for cust_type in customer_types]

# Perform ANOVA
f_statistic, p_value = f_oneway(*sales_by_customer_type)

print(f"ANOVA F-statistic for Customer Type: {f_statistic:.2f}")
print(f"ANOVA p-value for Customer Type: {p_value:.3f}")

if p_value < 0.05:
    print("\nConclusion: There are statistically significant differences in mean total sales across different customer types.")
else:
    print("\nConclusion: There are no statistically significant differences in mean total sales across different customer types.")
```

### 3. Linear Regression: Predicting 'Total' from 'Order Quantity'

We will build a simple linear regression model to predict 'Total' sales based on 'Order Quantity' and evaluate its statistical significance.

```python
import statsmodels.api as sm

# Define independent variable (X) and dependent variable (y)
X = df_cleaned['Order Quantity']
y = df_cleaned['Total']

# Add a constant to the independent variable for the intercept
X = sm.add_constant(X)

# Create and fit the OLS (Ordinary Least Squares) model
model = sm.OLS(y, X).fit()

# Print the model summary
print(model.summary())

print("\nInterpretation:")
print("The R-squared value indicates the proportion of variance in 'Total' sales explained by 'Order Quantity'.")
print("The p-value for 'Order Quantity' indicates its statistical significance. A p-value < 0.05 suggests a significant relationship.")
```

### 4. Multiple Linear Regression: Predicting 'Total' from Multiple Variables

Let's build a multiple linear regression model to predict 'Total' sales using several numerical features and evaluate its overall significance and the significance of individual predictors.

```python
import statsmodels.api as sm

# Define independent variables (X) and dependent variable (y)
X_multi = df_cleaned[['Cost Price', 'Retail Price', 'Profit Margin', 'Order Quantity', 'Discount $', 'Shipping Cost']]
y_multi = df_cleaned['Total']

# Add a constant to the independent variables for the intercept
X_multi = sm.add_constant(X_multi)

# Create and fit the OLS (Ordinary Least Squares) model
multi_model = sm.OLS(y_multi, X_multi).fit()

# Print the model summary
print(multi_model.summary())

print("\nInterpretation:")
print("This model evaluates the combined influence of several numerical factors on 'Total' sales.")
print("Look at the R-squared for overall fit and individual p-values for the significance of each predictor.")
```

### 5. Independent Samples t-test: Comparing 'Total' Sales Between Two Customer Types

Let's use an independent samples t-test to compare the mean 'Total' sales between two specific 'Customer Type' groups, for example, 'Corporate' and 'Consumer', to see if there's a statistically significant difference.

```python
from scipy.stats import ttest_ind

# Extract 'Total' sales for two specific customer types
sales_corporate = df_cleaned[df_cleaned['Customer Type'] == 'Corporate']['Total']
sales_consumer = df_cleaned[df_cleaned['Customer Type'] == 'Consumer']['Total']

# Perform independent samples t-test
t_statistic, p_value = ttest_ind(sales_corporate, sales_consumer, equal_var=False) # Assuming unequal variances, common with real-world data

print(f"T-statistic for Corporate vs. Consumer Total Sales: {t_statistic:.2f}")
print(f"P-value for Corporate vs. Consumer Total Sales: {p_value:.3f}")

if p_value < 0.05:
    print("\nConclusion: There is a statistically significant difference in mean total sales between Corporate and Consumer customer types.")
else:
    print("\nConclusion: There is no statistically significant difference in mean total sales between Corporate and Consumer customer types.")
```
